## Kalman Iteration

* **Kalman Filter**: An optimal linear algorithm that updates the expected value of a hidden variable based on the latest value of an observable variable.

  -In pairs trading, we need to dynamically estimate hidden variable _beta_ as an optimal hedge ratio, where *x(t)* and *y(t)* are prices and *e(t)* is a spread (Gaussian random noise) at given time *t*. The linear function that relates *y* and _beta_ is _y(t) = x(t)*beta(t) + e(t)_



* **Variables and Matrices**
 - Observable variable (vector): one of the price series *y* (n by 1)
 - Observation model (matrix): the other price series *x* + add constant (n by 2)
 - Hidden variable (vector): hedge ratio _beta_ (2 by 1), whose state covariance matrix is *R* (2 by 2)
 - State transition model (matrix): Identity matrix
   * beta at time t is the same as time t-1 plus noise
   * _beta(t) = beta(t-1) + w(t-1)_



* **State & Measurement Prediction**
    * State prediction: beta(t|t-1) = beta(t-1|t-1)
    * State covariance prediction: R(t|t-1) = R(t-1|t-1) + Vw
    * Measurement prediction: yhat(t) = x(t)beta(t|t-1)
    * Measurement variance prediction: Q(t) = x(t)'R(t|t-1)x(t) + Ve
   


* **State Updating**
    * beta(t|t) = beta(t|t-1) + K(t)*e(t)
    * R(t|t) = R(t|t-1) - K(t)*x(t)*R(t|t-1)
    * K(t) = R(t|t-1)*x(t)/Q(t)    (_Kalman gain_)
  
  
  
* **Initial conditions**
    * beta(1|0) = 0, R(0|0) = 0
    * Vw = delta/(1-delta)*Identity_matrix, where delta = 0.0001
    * Ve = 0.001

## Strategy

* Essentially mean-reverting strategy; long/short of spread e(t), which is forcast error for y(t) with given observation y(t-1).
* entry/exit position is given by standart deviation of e(t), which equals sqrt(Q(t))
* long when e(t) < sqrt(Q(t))
* short when e(t) > sqrt(Q(t))


````python
import numpy as np
import pandas as pd
import statsmodels.tsa.stattools as ts
import matplotlib.pyplot as plt

# load dataset
df_pairs = pd.read_csv('pair_list.csv')
df_pairs.drop(['Unnamed: 0'], axis = 1, inplace=True)
etfs = pd.read_csv('ETFs_merged.csv')
etfs['Date'] = pd.to_datetime(etfs['Date'])
etfs.drop(['Unnamed: 0'], axis = 1, inplace=True)
etfs.set_index('Date', inplace = True)

# Need to be avoided from warning message
pd.options.mode.chained_assignment = None

# Define summary tables
pair_list = [df_pairs.iloc[i][1]+' vs '+df_pairs.iloc[i][0] for i in range(len(df_pairs))]
output_list = ['APR (%)', 'num_of_trades/month', 'hit_ratio', 'std_returns', 'sharpe_ratio', 'MDD (%)']

df_summary = pd.DataFrame(index=pair_list, columns=output_list)
df_summary_cumret = pd.DataFrame(index = etfs.index[2:], columns=pair_list)
df_summary_drawdown =  pd.DataFrame(index = etfs.index[2:], columns=pair_list)

# Run Kalman iteration
for i in range(len(df_pairs)):
    pair_1 = df_pairs.iloc[i][1]
    pair_2 = df_pairs.iloc[i][0]
    pair_data = etfs[[pair_1,pair_2]]

    # Fist 2 datapoints should be removed because of Kalman filter iteration result
    df = etfs[[pair_1,pair_2]].iloc[2:]
    ret = df.pct_change()
    x = etfs[pair_1]
    y = etfs[pair_2]


    x=np.array(ts.add_constant(x))[:,[1,0]]

    delta=0.0001

    yhat=np.full(y.shape[0], np.nan)
    e=yhat.copy()
    Q=yhat.copy()

    # For clarity, we denote R(t|t) by P(t). Initialize R, P and beta.
    R=np.zeros((2,2))
    P=R.copy()
    beta=np.full((2, x.shape[0]), np.nan)
    Vw=delta/(1-delta)*np.eye(2)
    Ve=0.001
    beta[:, 0]=0

    for t in range(len(y)):
        if t > 0:
            beta[:, t]=beta[:, t-1]
            R=P+Vw

        yhat[t]=np.dot(x[t, :], beta[:, t])
    #     print('FIRST: yhat[t]=', yhat[t])

        Q[t]=np.dot(np.dot(x[t, :], R), x[t, :].T)+Ve
    #     print('Q[t]=', Q[t])

        # Observe y(t)
        e[t]=y.values[t]-yhat[t] # measurement prediction error
    #     print('e[t]=', e[t])
    #     print('SECOND: yhat[t]=', yhat[t])


        K=np.dot(R, x[t, :].T)/Q[t] #  Kalman gain
    #     print(K)

        beta[:, t]=beta[:, t]+np.dot(K, e[t]) #  State update
    #     print(beta[:, t])

        P=R-np.dot(np.dot(K, x[t, :]), R) # State covariance update
    #     print(R)

    pair_data['yhat'] = yhat
    pair_data['beta'] = beta[0,:]
    pair_data['spread'] = e
    pair_data['std'] = np.sqrt(Q)

    # Fist 2 datapoints should be removed because of Kalman filter iteration result
    pair_data = pair_data[pair_data.yhat !=0]

    # Entry/Exit condition
    pair_data['longsEntry'] = pair_data['spread']< -pair_data['std']
    pair_data['longsExit'] = pair_data['spread'] > 0
    pair_data['shortsEntry'] = pair_data['spread']> pair_data['std']
    pair_data['shortsExit'] = pair_data['spread'] < 0


    # Identifying position
    pair_data['positionsLong']  = np.nan
    pair_data['positionsShort']  = np.nan

    pair_data['positionsLong'][0]  = 0
    pair_data['positionsShort'][0]  = 0

    pair_data.loc[pair_data.longsEntry, 'positionsLong'] = 1
    pair_data.loc[pair_data.longsExit, 'positionsLong'] = 0
    pair_data.positionsLong = pair_data.positionsLong.fillna(method='ffill') 

    pair_data.loc[pair_data.shortsEntry, 'positionsShort'] = -1
    pair_data.loc[pair_data.shortsExit, 'positionsShort'] = 0
    pair_data.positionsShort = pair_data.positionsShort.fillna(method='ffill') 

    pair_data['positions'] = pair_data.positionsLong + pair_data.positionsShort


    pair_data['positions_size_x'] = pair_data.positions * (-pair_data.beta) * pair_data[df_pairs.iloc[i][1]]
    pair_data['positions_size_y'] = pair_data.positions * pair_data[df_pairs.iloc[i][0]]

    # Backtesting
    pair_data['pnl'] = pair_data.positions_size_x.shift(1) * ret.values[:,0] + pair_data.positions_size_y.shift(1) * ret.values[:,1]
    pair_data['returns'] = pair_data.pnl/(abs(pair_data.positions_size_x.shift())+abs(pair_data.positions_size_y.shift()))

    pair_data.pnl = pair_data.pnl.fillna(0)
    pair_data.returns = pair_data.returns.fillna(0)

    pair_data['cumpnl'] = pair_data.pnl.cumsum()
    pair_data['cumret'] = (1+pair_data.returns).cumprod()


    # Analysing Backtest Output
    APR = ((pair_data.cumret[-1])**(252/len(pair_data.cumret))-1)
    num_of_trades_per_month = len(pair_data[pair_data.positions != pair_data.positions.shift(1)]) * 252 / len(pair_data) /12
    hit_ratio = len(pair_data.pnl[pair_data.pnl>0])/len(pair_data.pnl[pair_data.pnl!=0])
    std_returns = pair_data.returns.std()
    sharpe_ratio = np.sqrt(252)*pair_data.returns.mean()/pair_data.returns.std()
    Daily_DD = (pair_data['cumret']/pair_data['cumret'].rolling(252,1).max() -1)
    Max_Daily_DD = Daily_DD.rolling(252,1).min()
    MDD = abs(min(Max_Daily_DD))

    # Updating summary tables
    df_summary.iloc[i] = [APR*100, num_of_trades_per_month, hit_ratio, std_returns, sharpe_ratio, MDD*100]
    df_summary_cumret[pair_list[i]] = pair_data.cumret
    df_summary_drawdown[pair_list[i]] = Max_Daily_DD
````
